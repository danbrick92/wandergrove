```python
from pathlib import Path 
import geopandas as gpd
import os
```

```python
# Establish pathing
data_dir = Path.joinpath(Path.cwd().parent, "data")
natural_earth_dir = Path.joinpath(data_dir, "natural_earth")
small_natural_earth_dir = Path.joinpath(natural_earth_dir, "1_to_50m_physical")
```

```python
for file in os.listdir(small_natural_earth_dir):
    if file.endswith(".shp"):
        print(file)
        file_path = Path.joinpath(small_natural_earth_dir, file)
        x = gpd.read_file(file_path)
        x.plot()

# geopandas.load()
```

```python
import rasterio

# Path to your raster file (can be .tif, .img, etc.)
file_path = Path.joinpath(natural_earth_dir, "NE1_HR_LC_SR_W_DR", "NE1_HR_LC_SR_W_DR.tif")
# Open the raster file
with rasterio.open(file_path) as src:
    raster_data = src.read(1)  # Read the first band of the raster
    print(f"Raster Shape: {raster_data.shape}")
    print(f"Raster CRS (Coordinate Reference System): {src.crs}")
    print(f"Raster Bounds: {src.bounds}")
```

```python
import matplotlib.pyplot as plt

# Visualize the raster using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(raster_data, cmap='viridis')
plt.colorbar(label='Pixel Value')
plt.title('Raster Visualization')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
